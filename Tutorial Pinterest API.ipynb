    <strong>redirect_URI:</strong> √à un URL che la tua applicazione fornisce a Pinterest per ricevere il codice di autorizzazione dopo che l'utente ha effettuato l'accesso e ha concesso i permessi alla tua app. ‚Üí <strong>Come impostare il Redirect_URI?</strong> <a href="https://docs.google.com/document/d/1oDvw2d0APmuqfhFQcACX4ljyptk33z7eyv8Yk6lR1v8/edit?usp=sharing" 
     style="color: white; background-color: red; text-decoration: none; padding: 2px 6px; border-radius: 4px;">Scoprilo qui</a>.
  </li>
  <li>
    <strong>response_type:</strong> Deve essere impostato su "code" per ottenere un codice di autorizzazione. Questo indica a Pinterest che questo √® l'inizio di una richiesta OAuth.
  </li>
  <li>
    <strong>scope:</strong> Definisce i permessi richiesti, ognuno di esso deve essere separato dall'altro con l‚Äôutilizzo di una ‚Äú,‚Äù. Inoltre, √® importante richiedere solo gli ambiti necessari per il funzionamento della tua applicazione, poich√© richiedere permessi non essenziali potrebbe influire negativamente sull'esperienza dell'utente e sulla probabilit√† che l'utente conceda l'autorizzazione. <a href="https://docs.google.com/document/d/1xHGZi6Q4VHLKbb4HLboOj3cugJXIDW68EH4oDPwtkJA/edit?usp=sharing" 
     style="color: white; background-color: red; text-decoration: none; padding: 2px 6px; border-radius: 4px;">Leggi qui</a> gli ambiti che √® possibile richiedere.
  </li>
  <li><strong>state:</strong> Un valore opzionale per prevenire attacchi CSRF.</li>
</ul>

<h5><strong>Esempio di URL che verr√† generato:</strong></h5>

<p style="text-align: justify;">
Per consentire all'utente di autenticarsi e concedere i permessi alla tua applicazione, devi reindirizzarlo al seguente URL, sostituendo <strong>{CLIENT_ID}</strong> e <strong>{REDIRECT_URI}</strong> con i valori corretti della tua app:
</p>

<p style="text-align: center; background-color: #f5f5f5; padding: 10px; border-radius: 5px;">
<code>https://www.pinterest.com/oauth/?client_id={CLIENT_ID}&redirect_url={REDIRECT_URI}&response_type=code&scope=boards:read,pins:read,pins:write</code>
</p>

```python
import webbrowser #Questa √® una libreria standard di Python che permette di aprire automatcamente un URL nel browser predefinito.
import json
import requests
from config import CLIENT_ID, REDIRECT_URI #Importiamo cos√¨ le credenziali salvate nel file di configurazione.

AUTH_URL = "https://www.pinterest.com/oauth/"
JSON_URL = "https://scuciecuci.altervista.org/oauth/auth_code.json" #URL del file JSON su Altervista, questo √® il link che possiamo prendere dalla cartella nel File manager di WP, ad esempio quando gi√† abbiamo generato il codice di autorizzazione con i permessi che vogliamo e non ne vogliamo generare altri.
LOCAL_JSON_PATH = "auth_code_local.json"  #File locale dove salvare il codice

def get_auth_code():
    #Controlla se il file JSON con il codice di autorizzazione √® disponibile online.
    #Se non esiste o non contiene il codice, avvia il processo di autenticazione.
    try:
        response = requests.get(JSON_URL)
        response.raise_for_status()  #Nel caso in cui la richiesta fallisce, solleva un errore.
        data = response.json()
        auth_code = data.get("auth_code")

        if auth_code: #Salviamo il codice in un file locale JSON
            with open(LOCAL_JSON_PATH, "w") as f:
                json.dump({"auth_code": auth_code}, f, indent=4)

            print("Il codice di autorizzazione √® stato trovato ed √® stato salvato in un file JSON.")
            return auth_code
    except requests.exceptions.RequestException:
        pass  #ignora l'errore

    #Se il codice non √® stato trovato, avviamo l'autenticazione, ed √® nel seguente modo che costruiamo l'URL:
    params = {"client_id": CLIENT_ID,
        "redirect_uri": REDIRECT_URI,
        "response_type": "code",
        "scope": "boards:read,pins:read,user_accounts:read,ads:read",} #Definiamo cos√¨ i permessi (=scopi) che la nostra applicazione vuole richiedere all'utente.
    auth_request_url = f"{AUTH_URL}?{'&'.join([f'{k}={v}' for k, v in params.items()])}"
    print("Apri il seguente link nel browser e autorizza l'app:", auth_request_url)
    webbrowser.open(auth_request_url) #nel caso in cui il browser non si apre automaticamente, l'utente pu√≤ copiare e incollare manualmente il link.

    input("Dopo aver autorizzato l'accesso, premi INVIO...")

    #Adesso riproviamo a leggere il file JSON dopo l'autenticazione:
    try:
        response = requests.get(JSON_URL)
        response.raise_for_status()
        data = response.json()
        auth_code = data.get("auth_code")

        if auth_code:
            #Salviamo il codice in un file locale JSON
            with open(LOCAL_JSON_PATH, "w") as f:
                json.dump({"auth_code": auth_code}, f, indent=4)

            print("Il codice di autorizzazione √® stato ottenuto ed √® stato salvato in un file JSON.")
            return auth_code
        else:
            print("Errore: il file JSON non contiene il codice.")
            return None
    except requests.exceptions.RequestException as e:
        print("Errore nel recupero del file JSON dopo l'autenticazione:", e)
        return None

#Eseguiamo il primo step
auth_code = get_auth_code()
```

<h5><strong style="color:red">4. Scambio del codice con il token di access</strong></h5>
<p>
Ora che abbiamo ottenuto il <strong>codice di autorizzazione</strong> (<code>ABCD1234</code>) dopo la fase di reindirizzamento, dobbiamo scambiarlo con un <strong>token di accesso</strong>. Il token di accesso √® essenziale perch√© permette alla nostra applicazione di effettuare <strong>chiamate API</strong> autenticandosi come l'utente.
</p>

<p>
Per ottenere il token di accesso, dobbiamo inviare una <strong>richiesta HTTP POST</strong> all'endpoint di Pinterest, per ulteriori informazioni, consulta la <a href="https://developers.pinterest.com/docs/api/v5/oauth-token/"> documentazione ufficiale</a>.<br>
</p>

<p>
Usiamo <strong>POST</strong> per ottenere il token perch√© <strong>protegge le credenziali sensibili</strong> e segue le <strong>best practice</strong> delle API OAuth 2.0. √à <strong>sconsigliato usare GET</strong> per autenticazione o invio di <strong>dati privati</strong>.
</p>
<p><strong>Per ottenerlo, questo √® il codice:</strong></p>

```python
import base64
import os
from config import CLIENT_ID, CLIENT_SECRET, REDIRECT_URI

#URL per ottenere o rinnovare token da Pinterest
TOKEN_URL = "https://api.pinterest.com/v5/oauth/token"

# Percorsi file locali
LOCAL_JSON_PATH = "auth_code_local.json"     #Contiene il codice di autorizzazione 
ACCESS_TOKEN_FILE = "access_token.json"      #File dove si salvano access_token e refresh_token

#Carica auth_code dal file locale (una sola volta, dopo il redirect iniziale)
def load_auth_code():
    try:
        with open(LOCAL_JSON_PATH, "r") as f:
            return json.load(f).get("auth_code")
    except:
        return None

#Salva la risposta completa (token + altri dati) nel file locale
def save_token_data(data):
    with open(ACCESS_TOKEN_FILE, "w") as f:
        json.dump(data, f, indent=4)

#Richiede un access_token partendo dal codice di autorizzazione (valido solo una volta)
def get_access_token(auth_code):
    credentials = f"{CLIENT_ID}:{CLIENT_SECRET}"
    encoded_credentials = base64.b64encode(credentials.encode()).decode()

    headers = {
        "Authorization": f"Basic {encoded_credentials}",
        "Content-Type": "application/x-www-form-urlencoded"
    }

    payload = {
        "grant_type": "authorization_code",
        "code": auth_code,
        "redirect_uri": REDIRECT_URI,
        "continuous_refresh": "true"  # Chiediamo anche il refresh_token
    }

    response = requests.post(TOKEN_URL, headers=headers, data=payload)
    data = response.json()
    save_token_data(data)

    return data.get("access_token") if response.status_code == 200 else None

#Usa il refresh_token per ottenere un nuovo access_token (automatico e senza browser)
def refresh_access_token(refresh_token):
    credentials = f"{CLIENT_ID}:{CLIENT_SECRET}"
    encoded_credentials = base64.b64encode(credentials.encode()).decode()

    headers = {
        "Authorization": f"Basic {encoded_credentials}",
        "Content-Type": "application/x-www-form-urlencoded"
    }

    payload = {
        "grant_type": "refresh_token",
        "refresh_token": refresh_token
    }

    response = requests.post(TOKEN_URL, headers=headers, data=payload)
    data = response.json()
    save_token_data(data)

    return data.get("access_token") if response.status_code == 200 else None

#Prova a caricare token esistenti dal file locale (se gi√† presenti)
def load_existing_token():
    if not os.path.exists(ACCESS_TOKEN_FILE):
        return None, None
    try:
        with open(ACCESS_TOKEN_FILE, "r") as f:
            data = json.load(f)
            return data.get("access_token"), data.get("refresh_token")
    except:
        return None, None

#1. Carica access_token e refresh_token se gi√† salvati
access_token, refresh_token_value = load_existing_token()

#2. Se non c'√® access_token ma abbiamo un refresh_token ‚Üí prova a rinnovarlo
if not access_token and refresh_token_value:
    access_token = refresh_access_token(refresh_token_value)

#3. Se non abbiamo nemmeno quello ‚Üí usa auth_code salvato per ottenere token
if not access_token:
    auth_code = load_auth_code()
    if auth_code:
        access_token = get_access_token(auth_code)

#4. Verifica finale
if access_token:
    print("Access token pronto.")
else:
    print("Errore: token non ottenuto. Riprova il processo di autorizzazione.")
```

Questo script Python si occupa di gestire il <strong>flusso di autenticazione OAuth 2.0</strong> con Pinterest, in particolare per ottenere e rinnovare un <strong>access token</strong> necessario per effettuare chiamate API protette.
</p>

<p>
All‚Äôinizio, vengono definiti alcuni valori chiave:
</p>

<p>
Il codice contiene quindi alcune funzioni fondamentali:
</p>

<ul>
  <li><code><strong>load_auth_code()</strong></code>: legge dal file locale il codice di autorizzazione (<code>auth_code</code>). Questo codice √® valido solo una volta e serve per richiedere il token iniziale.</li>
  
  <li><code><strong>save_token_data(data)</strong></code>: salva la risposta JSON completa della richiesta token (che include access token, refresh token e altri dati) in un file locale per uso futuro.</li>
  
  <li><code><strong>get_access_token(auth_code)</strong></code>: utilizza il <code>auth_code</code> per ottenere un <strong>access token</strong> tramite una richiesta POST all‚Äôendpoint di Pinterest. Nella richiesta:</li>
  <ul>
    <li>Le credenziali client (<code>CLIENT_ID</code> e <code>CLIENT_SECRET</code>) sono codificate in Base64 e inserite nell‚Äôheader <code>Authorization</code> secondo lo standard OAuth 2.0.</li>
    <li>Il payload include il <code>grant_type</code> ‚Äúauthorization_code‚Äù, il codice di autorizzazione, l‚ÄôURI di redirect e il parametro <code>continuous_refresh</code> impostato a ‚Äútrue‚Äù per richiedere anche il refresh token.</li>
  </ul>
  
  <li><code><strong>refresh_access_token(refresh_token)</strong></code>: utilizza il <strong>refresh token</strong> salvato per ottenere un nuovo access token senza dover passare nuovamente dal browser o dal processo manuale di autorizzazione. Anche qui, le credenziali sono codificate in Base64 e inviate nell‚Äôheader, mentre nel payload il <code>grant_type</code> √® ‚Äúrefresh_token‚Äù.</li>
  
  <li><code><strong>load_existing_token()</strong></code>: tenta di leggere dal file locale eventuali token gi√† salvati, restituendo <code>access_token</code> e <code>refresh_token</code> se disponibili.</li>

Utilizzando l'<strong>API di Pinterest</strong>, faremo una <strong>richiesta</strong> per ottenere un <strong>file CSV</strong> contenente informazioni sull‚Äô<strong>ID dei pin</strong>, il <strong>titolo</strong>, la <strong>descrizione</strong>, la <strong>data di creazione</strong> e gli <strong>hashtag associati</strong>.

```python
import csv

API_URL = "https://api.pinterest.com/v5/boards/{board_id}/pins"
ACCESS_TOKEN_FILE = "access_token.json"
BOARD_ID = "1085578753869149364"  # l'ID board "Vari ed eventuali"

# Carica il token di accesso salvato
try:
    with open(ACCESS_TOKEN_FILE, "r") as f:
        data = json.load(f)
        access_token = data.get("access_token")

        if not access_token:
            print("ERRORE: Il file JSON non contiene un access token valido.")
            exit()
except FileNotFoundError:
    print(f"ERRORE: Il file {ACCESS_TOKEN_FILE} non esiste. Assicurati di aver ottenuto il token di accesso.")
    exit()

def get_pins(board_id, access_token): #Estrae tutti i Pin da una board, inclusa la data di creazione, gestendo la paginazione.
    pins = []
    headers = {"Authorization": f"Bearer {access_token}"}
    url = API_URL.format(board_id=board_id)
    params = {"page_size": 25, "fields": "id,title,description,created_at"}  # Includi il campo created_at
    bookmark = None  # Serve per la paginazione

    while True:
        if bookmark:
            params["bookmark"] = bookmark
        
        response = requests.get(url, headers=headers, params=params)

        if response.status_code != 200:
            print(f"ERRORE: {response.status_code} - {response.text}")
            break

        data = response.json()
        pins.extend(data.get("items", []))

        bookmark = data.get("bookmark")
        if not bookmark:
            break
    
    return pins

def save_pins_to_csv(pins, OUTPUT_FILE="pins_data.csv"):
    #Salva i dati dei Pin in un file CSV, inclusa la data di creazione.
    with open(OUTPUT_FILE, "w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow(["ID", "Titolo", "Descrizione", "Data di Creazione", "Numero Hashtag"])

        for pin in pins:
            pin_id = pin.get("id", "")
            title = pin.get("title", "")
            description = pin.get("description", "")
            created_at = pin.get("created_at", "")
            hashtags = description.count("#") if description else 0

            writer.writerow([pin_id, title, description, created_at, hashtags])

    print(f"Dati salvati in {OUTPUT_FILE}")

# Esegui il codice
pins = get_pins(BOARD_ID, access_token)

if pins:
    # Stampa i primi 5 pin
    print("üìå Primi 5 pin trovati:")
    for i, pin in enumerate(pins[:5], 1):
        print(f"\nPin {i}:")
        print(f"ID: {pin.get('id', '')}")
        print(f"Titolo: {pin.get('title', '')}")
        print(f"Descrizione: {pin.get('description', '')}")
        print(f"Data di creazione: {pin.get('created_at', '')}")
    
    save_pins_to_csv(pins)
else:
    print("Nessun Pin trovato nella board.")
```

<ol start="3">
  <li><strong>Pulizia del testo grezzo</strong><br>
  La <strong>normalizzazione del testo</strong>, attraverso la <strong>conversione in minuscolo</strong> e la <strong>rimozione di punteggiatura</strong>, numeri e spazi extra, √® un passaggio importante perch√© senza questa pulizia, lo stesso termine potrebbe essere interpretato diversamente (es. ‚Äú<strong>Cane</strong>‚Äù vs ‚Äú<strong>cane</strong>‚Äù) o confondere gli algoritmi con simboli non rilevanti per l‚Äô<strong>analisi semantica</strong>.</li>
</ol>

```python
df = pd.read_csv('Dataset_aggiornato.csv')

#Convertiamo il testo in minuscolo
df['text'] = df['text'].str.lower()

#La variabile string.punctuation include i seguenti caratteri standard: 
#!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~
#ma nelle descrizioni e nei titoli sono presenti anche trattini -- e i punti di sospensione ‚Ä¶ 
#che non sono in string.punctuation. Quindi:
def clean_punctuation(text):
    #Rimuoviamo la punteggiatura ASCII standard
    text = text.translate(str.maketrans('', '', string.punctuation))
    #Rimuoviamo i punti di sospensione (‚Ä¶)
    text = re.sub(r'‚Ä¶', '', text)
    #Rimuoviamo trattini lunghi e simili (‚Äî, ‚Äì)
    text = re.sub(r'[‚Äì‚Äî]', '', text)
    #Rimuoviamo triple punti "..."
    text = re.sub(r'\.\.\.', '', text)
    return text

#Applichiamo la funzione di pulizia della punteggiatura
df['text'] = df['text'].apply(clean_punctuation)

#Rimuoviamo i numeri
df['text'] = df['text'].str.replace(r'\d+', '', regex=True)

#Rimuoviamo gli spazi bianchi multipli
df['text'] = df['text'].str.strip()
df['text'] = df['text'].replace(r'\s+', ' ', regex=True)

#Salviamo:
df.to_csv('Dataset_aggiornato.csv', index=False)
```

```python
#Scarichiamo (se necessario)
try:
    nltk.data.find('corpora/stopwords')
except LookupError:
    nltk.download('stopwords')

#Carichiamo il dataset
df = pd.read_csv('Dataset_aggiornato.csv')

#La colonna tokens √® salvata come stringa, convertiamola in lista
df['tokens'] = df['tokens'].apply(ast.literal_eval)

#Definiamo il set di stopword in inglese
stop_words = set(stopwords.words('english'))

#Creiamo la nuova colonna 'no_stopwords'
df['no_stopwords'] = df['tokens'].apply(lambda tokens: [word for word in tokens if word.lower() not in stop_words])

#Verifichiamo
print(df[['tokens', 'no_stopwords']].head(3))

#Salviamo:
df.to_csv('Dataset_aggiornato.csv', index=False)
```

<ol start="6">
  <li><strong>Stemming</strong><br>
  Lo <strong>stemming</strong> riduce le parole alle loro <strong>radici</strong>, uniformando varianti morfologiche diverse di uno stesso <strong>lemma</strong> (es. ‚Äú<strong>correndo</strong>‚Äù, ‚Äú<strong>corre</strong>‚Äù ‚Üí ‚Äú<strong>corr</strong>‚Äù). Questa operazione facilita il riconoscimento di <strong>concetti simili</strong>, migliorando la <strong>performance</strong> degli algoritmi di <strong>analisi</strong> e <strong>classificazione</strong>.</li>
</ol>


```python
df = pd.read_csv('Dataset_aggiornato.csv')

#La colonna 'no_stopwords' √® salvata come stringa, convertiamola in lista
df['no_stopwords'] = df['no_stopwords'].apply(ast.literal_eval)

#Inizializziamo lo stemmer (PorterStemmer per l'inglese)
stemmer = PorterStemmer()

#Creiamo la nuova colonna 'stemmed' applicando lo stemming ai token senza stopword
df['stemmed'] = df['no_stopwords'].apply(lambda tokens: [stemmer.stem(word) for word in tokens])

#Verifichiamo
print(df[['no_stopwords', 'stemmed']].iloc[5])

#Salviamo il dataset aggiornato:
df.to_csv('Dataset_aggiornato.csv', index=False)
```

<h5><strong style="color:red">Analisi statistica della distribuzione di <code>repin_count</code></strong></h5>

<p>Adesso ci concentreremo sull‚Äô<strong>analisi statistica</strong> della distribuzione dei valori di <code>repin_count</code> per identificare eventuali <strong>valori anomali</strong> (<strong>outlier</strong>) che potrebbero distorcere l‚Äôinterpretazione complessiva.</p>

<p>In generale noi abbiamo <strong>4.999 post</strong>. Osservando la distribuzione, notiamo:</p>
<ul>
  <li>Molti post hanno pochi repin (0‚Äì10): distribuzione fortemente <strong>sbilanciata a sinistra</strong>.</li>
  <li>Solo in pochi hanno <strong>repin &gt; 1000</strong>, fino a un massimo di <strong>2053</strong>.</li>
</ul>

<p>Utilizzeremo il metodo dell‚Äô<strong>Interquartile Range (IQR)</strong> per definire una <strong>soglia</strong> oltre la quale considerare i post come <strong>estremamente popolari</strong> e potenzialmente anomali.</p>

<p>Successivamente, suddivideremo il dataset, basandoci su questa soglia e sulla <strong>mediana</strong> dei valori, in modo da avere una classificazione pi√π chiara e significativa, in due gruppi principali:</p>
<ul>
  <li><strong>Post meno popolari</strong></li>
  <li><strong>Post pi√π popolari</strong></li>
</ul>


```python
df = pd.read_csv('Dataset_aggiornato.csv')

#Calcoliamo i quartili e IQR
Q1 = df['repin_count'].quantile(0.25)
Q3 = df['repin_count'].quantile(0.75)
IQR = Q3 - Q1
soglia_outlier = Q3 + 1.5 * IQR

#Isoliamo gli outlier e salviamoli
outlier = df[df['repin_count'] > soglia_outlier]
outlier.to_csv('outlier_repin.csv', index=False)

#Rimuoviamo gli outlier dal dataset principale
df_senza_outlier = df[df['repin_count'] <= soglia_outlier]

#Stampiamo un riepilogo:
print(f"Totale post: {len(df)}")
print(f"Post senza outlier: {len(df_senza_outlier)}")
print(f"Post salvati come outlier: {len(outlier)}")
print(f"Soglia usata per individuare gli outlier: {soglia_outlier}")
```

```python
#Calcoliamo la mediana sui dati filtrati
mediana_repin = df_senza_outlier['repin_count'].median()
print(f"\nMediana dei repin (senza outlier): {mediana_repin}")

#Dividiamo in due gruppi:
gruppo_popolari = df_senza_outlier[df_senza_outlier['repin_count'] >= mediana_repin]
gruppo_meno_popolari = df_senza_outlier[df_senza_outlier['repin_count'] < mediana_repin]

print(f"\nPost pi√π popolari (>= mediana): {len(gruppo_popolari)}")
print(f"Post meno popolari (< mediana): {len(gruppo_meno_popolari)}")
```

```python
#Funzione per ottenere le frequenze delle parole usando CountVectorizer
def ottieni_frequenze(df_group, max_features=100):
    vectorizer = CountVectorizer(max_features=max_features)
    X = vectorizer.fit_transform(df_group['stemmed'].fillna(''))
    frequenze = X.sum(axis=0).A1
    parole = vectorizer.get_feature_names_out()
    df_frequenze = pd.DataFrame({'Parola': parole, 'Frequenza': frequenze})
    return df_frequenze.sort_values(by='Frequenza', ascending=False).head(20)

#Otteniamo le 20 parole pi√π frequenti per ciascun gruppo
df_top_popolari = ottieni_frequenze(gruppo_popolari)
df_top_meno_popolari = ottieni_frequenze(gruppo_meno_popolari)

plt.figure(figsize=(14, 6))

#Post pi√π popolari
plt.subplot(1, 2, 1)
sns.barplot(
    y='Parola',
    x='Frequenza',
    hue='Parola',
    data=df_top_popolari,
    palette='Blues_d',
    legend=False
)
plt.title('Parole pi√π frequenti - Post pi√π popolari (esclusi outlier)')

#Post meno popolari
plt.subplot(1, 2, 2)
sns.barplot(
    y='Parola',
    x='Frequenza',
    hue='Parola',
    data=df_top_meno_popolari,
    palette='Greens_d',
    legend=False
)
plt.title('Parole pi√π frequenti - Post meno popolari')

plt.tight_layout()
plt.show()
```

```python
#Estraiamo le parole del primo grafico nell'ordine corretto (dal pi√π frequente al meno frequente)
parole_popolari_ordinate = df_top_popolari['Parola'].tolist()

#Creiamo un dizionario delle frequenze per i post meno popolari
freq_meno_popolari_dict = dict(zip(df_top_meno_popolari['Parola'], df_top_meno_popolari['Frequenza']))

#Adesso creiamo la lista delle frequenze per i post meno popolari seguendo l'ordine del primo grafico
frequenze_meno_pop_ordinate = []
for parola in parole_popolari_ordinate:
    if parola in freq_meno_popolari_dict:
        frequenze_meno_pop_ordinate.append(freq_meno_popolari_dict[parola])
    else:
        frequenze_meno_pop_ordinate.append(0)

#Poi generiamo il DataFrame per il secondo grafico con l'ordine corretto
df_meno_pop_ordinato = pd.DataFrame({
    'Parola': parole_popolari_ordinate,
    'Frequenza': frequenze_meno_pop_ordinate
})

#Visualizziamo i grafici
plt.figure(figsize=(14, 6))

#Post pi√π popolari (grafico originale)
plt.subplot(1, 2, 1)
sns.barplot(
    y='Parola',
    x='Frequenza',
    hue='Parola',
    data=df_top_popolari,
    palette='Blues_d',
    legend=False
)
plt.title('Parole pi√π frequenti - Post pi√π popolari (esclusi outlier)')

#Post meno popolari (con ordine delle parole uguale al primo grafico)
plt.subplot(1, 2, 2)
sns.barplot(
    y='Parola',
    x='Frequenza',
    hue='Parola',
    data=df_meno_pop_ordinato,
    palette='Greens_d',
    legend=False
)
plt.title('Parole pi√π frequenti - Post meno popolari (stesso ordine)')

plt.tight_layout()
plt.show()

#Stampiamo le statistiche per verifica
print("\nConfronto frequenze - Parole ordinate come nel primo grafico:")
print("=" * 65)
print(f"{'Pos':<3} {'Parola':<15} {'Pop. Posts':<12} {'Meno Pop.':<12} {'Differenza':<12}")
print("-" * 65)

for i, parola in enumerate(parole_popolari_ordinate):
    freq_pop = df_top_popolari[df_top_popolari['Parola'] == parola]['Frequenza'].iloc[0]
    freq_meno_pop = frequenze_meno_pop_ordinate[i]
    differenza = freq_pop - freq_meno_pop
    print(f"{i+1:<3} {parola:<15} {freq_pop:<12} {freq_meno_pop:<12} {differenza:<12}")

print("-" * 65)
print(f"Parole con freq. 0 nei post meno popolari: {sum(1 for x in frequenze_meno_pop_ordinate if x == 0)}")
print(f"Parole presenti in entrambi i gruppi: {sum(1 for x in frequenze_meno_pop_ordinate if x > 0)}")
```

Tuttavia, potrebbero emergere <strong>sfumature interessanti</strong> nelle frequenze relative e nella presenza di termini specifici che caratterizzano i contenuti.</p>

```python
def calcola_frequenze(df_group, colonna_testo='stemmed', top_n=20):
    vectorizer = CountVectorizer()
    X = vectorizer.fit_transform(df_group[colonna_testo].dropna())
    parole = vectorizer.get_feature_names_out()
    frequenze = X.sum(axis=0).A1
    df_freq = pd.DataFrame({'Parola': parole, 'Frequenza': frequenze})
    df_freq = df_freq.sort_values(by='Frequenza', ascending=False).head(top_n)
    return df_freq

freq_popolari_aggiornati = calcola_frequenze(gruppo_popolari_aggiornato)
freq_meno_popolari = calcola_frequenze(gruppo_meno_popolari)

plt.figure(figsize=(16, 7))

plt.subplot(1, 2, 1)
sns.barplot(
    data=freq_popolari_aggiornati,
    y='Parola',
    x='Frequenza',
    hue='Parola',           
    palette='Blues_d',
    dodge=False,
    legend=False           
)
plt.title('Parole pi√π frequenti - Post pi√π popolari (inclusi outlier)')
plt.xlabel('Frequenza')
plt.ylabel('Parola')

plt.subplot(1, 2, 2)
sns.barplot(
    data=freq_meno_popolari,
    y='Parola',
    x='Frequenza',
    hue='Parola',          
    palette='Greens_d',
    dodge=False,
    legend=False           
)
plt.title('Parole pi√π frequenti - Post meno popolari')
plt.xlabel('Frequenza')
plt.ylabel('')

plt.tight_layout()
plt.show()
```

Tra questi, spiccano parole come:</p>
